```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx.readwrite import json_graph
import json

# Load the graph data from the JSON file
with open("data/graph_snapshot.json", "r") as f:
    data = json.load(f)

# Recreate the NetworkX graph
G = json_graph.node_link_graph(data)

# Define colors for different node types
node_colors = {
    "Program": "skyblue",
    "Org": "lightcoral",
    "Criterion": "lightgreen",
    "DocumentType": "gold"
}

# Assign colors to nodes based on their type attribute
color_map = [node_colors.get(G.nodes[node]['type'], 'gray') for node in G.nodes()]

# Prepare labels for nodes (using 'name' for Programs/Orgs, 'id' for others)
node_labels = {}
for node_id, attributes in G.nodes(data=True):
    if 'name' in attributes:
        node_labels[node_id] = attributes['name']
    else:
        node_labels[node_id] = node_id

plt.figure(figsize=(18, 12)) # Increase figure size for better visibility
pos = nx.spring_layout(G, k=0.15, iterations=50) # Use spring layout for better node distribution
nx.draw_networkx_nodes(G, pos, node_color=color_map, node_size=2000)
nx.draw_networkx_edges(G, pos, alpha=0.5, arrowsize=20)
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=8, font_weight='bold')
plt.title("Knowledge Graph of Programs, Organizations, Criteria, and Documents", size=16)
plt.axis('off') # Hide axes
plt.tight_layout()
plt.show()
```
